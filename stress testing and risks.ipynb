```python
myPortfolio = pd.read_excel("OutstandingLoans.xlsx")

myLoanHistory = pd.read_excel("OldLoans.xlsx")

CBHistory = pd.read_csv("Simplified_History.csv")

combinedHistory = pd.merge(myLoanHistory, CBHistory, on = "Year")
```

```python
def check_data_completeness(data):
    missing_data = data.isnull().sum()
    if missing_data.sum() == 0:
        print("Все столбцы данных заполнены равномерно.")
    else:
        print("Есть пропущенные значения в следующих столбцах:")
        print(missing_data[missing_data > 0])


check_data_completeness(combinedHistory)
```

Мы получаем таблицу для проведения стресс-тестирования. Здесь указаны следующие колонки:
* RiskLevel: Этот столбец обозначает уровень риска, связанный с клиентом или финансовым продуктом. Обычно уровни риска классифицируются как низкий, средний или высокий, что помогает в оценке вероятности возникновения финансовых потерь.
* GDP: ВВП страны
* DJX_Return: Исторические данные по Dow Jones Industrial Average
* Default: Невыполнение обязательств по кредиту. Этот столбец содержит бинарные данные, где 1 означает, что клиент не выполнил обязательства по кредиту, а 0 - что выполнил.
* YOB: Год обслуживания

Создается генерализованная линейная модель (GLM) с биномиальным распределением для оценки вероятности дефолта (Default) в зависимости от уровня риска (RiskLevel), года начала обслуживания (YOB), доходности DJX (DJX_Return) и ВВП (GDP).

```python
model = smf.glm(formula = "Default ~ RiskLevel + YOB + DJX_Return + GDP", 
                data = combinedHistory, 
                family = sm.families.Binomial())

result = model.fit()

print(result.summary())

predictions = result.predict()

#Рассчитывается историческая частота дефолтов по каждому году обслуживания (YOB). 
#Частота дефолтов определяется как процент дефолтов в каждом году.
Default = []
for i in range(np.shape(combinedHistory["YOB"].unique())[0]):
  YOB = combinedHistory.loc[combinedHistory["YOB"] == i + 1]
  DefaultFreq = YOB["Default"].value_counts()
  DefaultRate = (DefaultFreq.values[1]/(DefaultFreq.values[0] + DefaultFreq.values[1]))*100
  Default.append(DefaultRate)

#DataFrame с годами обслуживания и предсказанными вероятностями дефолта
myModel = pd.DataFrame({"YOB": np.array(combinedHistory["YOB"]), "estDefault": predictions})
# рассмотреть несбалансированность данных
#Рассчитываются предсказанные частоты дефолтов для каждого года обслуживания как среднее значение предсказанных вероятностей дефолта.
estDefault = []
for i in range(np.shape(myModel["YOB"].unique())[0]):
  YOB = myModel.loc[myModel["YOB"] == i + 1]
  estDefaultFreq = YOB["estDefault"].mean()*100
  estDefault.append(estDefaultFreq)

AdverseScenario = pd.read_csv("Simplified_Adverse.csv")

# Portfolio under adverse economic conditions.
AdversePortfolio = myPortfolio.assign(Year = AdverseScenario.iloc[0, 0], 
                                      DJX_Return = AdverseScenario.iloc[0, 1], 
                                      GDP = AdverseScenario.iloc[0, 2])

PD = result.predict(AdversePortfolio)

#DataFrame с годами обслуживания и предсказанными вероятностями дефолта в неблагоприятных условиях
predPD = pd.DataFrame({"YOB": np.array(myPortfolio["YOB"]), "PD": PD})

На этих двух графиках вы увидите сравнение трех различных оценок частот дефолтов:

Синий график: Исторические данные о частотах дефолтов.
Зеленый график: Частоты дефолтов, предсказанные моделью на основе имеющихся данных.
Красный график: Частоты дефолтов, предсказанные моделью в условиях неблагоприятного экономического сценария.
predDefault = []
for i in range(np.shape(predPD["YOB"].unique())[0]):
  YOB = predPD.loc[predPD["YOB"] == i + 1]
  predDefaultFreq = YOB["PD"].mean()*100
  predDefault.append(predDefaultFreq)

#Ожидаемая потеря вычисляется как сумма произведений экспозиции на момент дефолта (EAD), 
# коэффициента потерь при дефолте (LGD) и предсказанных вероятностей дефолта для каждой записи в портфеле.
ExpectedLoss = sum(AdversePortfolio["EAD"]*AdversePortfolio["LGD"]*PD)

plt.subplot(2, 1, 1)
plt.plot(combinedHistory["YOB"].unique(), Default, "o-", color = "blue", label = "Histotical Portfolio")
plt.title("Defaul Rate (%)")
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(myModel["YOB"].unique(), estDefault, "o-", color = "green", label = "Fitted Model")
plt.xlabel("Years on Books")
plt.legend()

plt.plot(predPD["YOB"].unique(), predDefault, "o-", color = "red", label = "Predicted Adverse Portfolio")
plt.title("Defaul Rate (%)")
plt.xlabel("Years on Books")
plt.legend()

plt.show()
```

На этих двух графиках вы увидите **сравнение** трех различных оценок частот дефолтов:

* Синий график: Исторические данные о частотах дефолтов.
* Зеленый график: Частоты дефолтов, предсказанные моделью на основе имеющихся данных.
* Красный график: Частоты дефолтов, предсказанные моделью в условиях неблагоприятного экономического сценария.

**Value at Risk (VAR)** — это статистическая мера риска, которая показывает максимальные потери портфеля за определенный период времени при заданном уровне доверия. VAR измеряет вероятность того, что потери портфеля превысят определенную сумму за определенный период времени при заданной вероятности.

ES оценивает ожидаемую величину потерь, превышающих значение VAR, и таким образом предоставляет информацию о "хвостовых рисках" или экстремальных потерях.

### 1. Historical simulation

Допущение: **будущие** доходы будут распределяться **аналогично** историческим.

Таким образом, для доходности портфеля нет **конкретного распределения**.

```python
# задаем начальную стоимость портфеля
initial_portfolio = 100000

# процентное изменение цены по сравнению с предыдущим периодом
returns = df.pct_change()

# вычисляем доходность портфеля как средневзвешенную доходность отдельных активов
weights = np.full((4), 0.25) # aпредполагаем равные веса
port_returns = (weights * returns).sum(axis=1) #  вычисляем доходность портфеля как взвешенную сумму доходностей отдельных активов по каждому дню

# мы задаем уровень доверия (95%) и вычисляем VaR
confidence_level = 0.95
# P(Return <= VAR) = alpha
# определяет значение VaR, такое что 95% времени доходность портфеля будет выше этого значения
var = port_returns.quantile(q=1-confidence_level)
# здесь мы вычисляем CVaR, что представляет собой среднее значение доходностей ниже уровня VaR
# мы фильтруем доходности, которые меньше или равны VaR, и затем вычисляем их среднее.
cvar = port_returns[port_returns <= var].mean()

# переводим VaR и CVaR из процентных значений в абсолютные денежные величины, умножая их на начальную стоимость портфеля
var_value = var * initial_portfolio
cvar_value = cvar * initial_portfolio

print(f"Historical VaR at {confidence_level} confidence level: {var_value:.2f} ({var:.2%})")
print(f"Historical CVaR at {confidence_level} confidence level: {cvar_value:.2f} ({cvar:.2%})")
```

```python
plt.hist(port_returns, bins=20, density=True, alpha=0.3)

plt.axvline(x=var, color='red', linestyle='--', label=f"VaR: {var:.2%}")
plt.axvline(x=cvar, color='blue', linestyle='--', label=f"CVaR: {cvar:.2%}")

plt.legend()
plt.xlabel("Доходность портфеля")
plt.ylabel("Плотность вероятности")
plt.title(f"Процентная доходность портфеля")
plt.show()
```

### 2. Параметрический метод

Допущение 1: доходность соответствует **нормальному** распределению

Допущение 2: **нормальность**

Стоит сразу пояснить разницу между 1 и 2 допущением

* Допущение 1:
Мы предполагаем, что доходности нашего портфеля строго следуют нормальному распределению. Поэтому мы используем точные параметры нормального распределения (например, среднее и стандартное отклонение) для расчетов VaR и CVaR.

* Допущение 2:
Доходности нашего портфеля могут быть аппроксимированы нормальным распределением, хотя в реальности они могут иметь некоторые отклонения. Мы используем нормальное распределение для упрощения расчетов, зная, что это приближение, а не точное соответствие.


Рассчитываем VaR на основе z-score, соответствующего доверительному уровню (95%)

Рассчитываем CVAR на основе предположения о нормальности.

```python

initial_portfolio = 100000

returns = df.pct_change()

# вычисление среднего и ковариационной матрицы доходностей
mean_returns = returns.mean()
cov_matrix = returns.cov() # показывает, как доходности различных акций в портфеле изменяются вместе


weights = np.full((4), 0.25)
# вычисляем среднюю доходность портфеля как взвешенную сумму средних доходностей отдельных акций
port_mean_return = (weights * mean_returns).sum()
# стандартное отклонение доходности портфеля с использованием ковариационной матрицы
# используется матричное умножение весов на ковариационную матрицу и обратно на веса
port_std_dev = np.sqrt(weights.T @ cov_matrix @ weights) 
 
confidence_level = 0.95
# z-оценка для заданного уровня доверия, что соответствует квантилю стандартного нормального распределения
z_score = norm.ppf(q=1-confidence_level)
# учитываем стандартное отклонение и среднюю доходность портфеля
# мы используем отрицательное значение, так как VaR представляет собой убыток
var = - (norm.ppf(confidence_level)*port_std_dev - port_mean_return)

# эта формула учитывает среднюю доходность, стандартное отклонение и значение плотности нормального распределения для z-оценки
cvar = 1 * (port_mean_return - port_std_dev * (norm.pdf(z_score)/(1-confidence_level)))

var_initial = initial_portfolio * var
cvar_initial = initial_portfolio * cvar

print(f"Parametric VaR at {confidence_level} confidence level: {var_initial:.2f} ({var:.2%})")
print(f"Parametric CVaR at {confidence_level} confidence level: {cvar_initial:.2f} ({cvar:.2%})")
```

```python
plt.hist(port_returns, bins=20, density=True, alpha=0.3)

plt.axvline(x=var, color='red', linestyle='--', label=f"VaR: {var:.2%}")
plt.axvline(x=cvar, color='blue', linestyle='--', label=f"CVaR: {cvar:.2%}")

plt.legend()
plt.xlabel("Доходность портфеля")
plt.ylabel("Плотность вероятности")
plt.title(f"Процентные доходности портфеля")
plt.show()
```

### 3. Метод Монте-Карло

**Допущение**: будущие доходы портфеля следуют многомерному нормальному распределению, которое в реальности может не соответствовать действительности

Перед тем, как посмотреть на код, стоит описать некоторые нюансы при реализации

* **Разложение Холецкого**


Разложение Холецкого используется для генерации коррелированных случайных переменных. В контексте моделирования доходностей акций, мы хотим учитывать корреляции между доходностями различных акций. Разложение Холецкого помогает преобразовать набор независимых случайных переменных в набор коррелированных случайных переменных, соответствующих заданной ковариационной матрице.

Это разложение позволит нам создать набор случайных чисел с такой же структурой корреляций, как у оригинальных данных.
Если 𝑍 — матрица случайных чисел, сгенерированных из стандартного нормального распределения, то 𝐿⋅𝑍 будет матрицей коррелированных случайных чисел с ковариационной матрицей 𝐴.

```python
# устанавливаем случайное зерно для воспроизводимости
np.random.seed(123)

initialPortfolio = 100000

n_simulation = 400  # количество симуляций

T = 252  # количество торговых дней в году

weights = np.full((4), 0.25)

# средняя доходность, повторенная для каждого торгового дня
meanM = np.full(shape=(T, len(weights)), fill_value=mean_returns).T

# инициализируем матрицу для хранения процентного изменения для каждой симуляции
sim_pct_change = np.full(shape=(T, n_simulation), fill_value=0.0)

for m in range(n_simulation):
    # генерируем матрицу случайных чисел для каждого дня в периоде T и каждой акции в портфеле
    Z = np.random.normal(size=(T, len(weights)))
    
    # получаем разложение Холецкого ковариационной матрицы
    L = np.linalg.cholesky(cov_matrix)
    
    # вычисляем ежедневное процентное изменение с использованием разложения Холецкого
    daily_pct_change = meanM + np.inner(L, Z) #скалярное произведение массивов
    
    # вычисляем смоделированное процентное изменение портфеля для каждого дня в периоде T
    sim_pct_change[:, m] = np.inner(weights, daily_pct_change.T)
```

```python
plt.plot(sim_pct_change)

plt.axhline(np.percentile(sim_pct_change,5), color='r', linestyle='dashed', linewidth=1, label='5th Percentile')
plt.axhline(np.percentile(sim_pct_change,95), color='g', linestyle='dashed', linewidth=1,  label='95th Percentile')
plt.axhline(np.mean(sim_pct_change), color='b', linestyle='solid', linewidth=1, label='Mean')

plt.legend()
plt.ylabel('Изменение стоимости портфеля в процентах')
plt.xlabel('Дни')
plt.title('MК симуляции процентного изменения портфеля')
plt.show()
```

5-й процентиль:
* Если симулированное изменение стоимости портфеля находится ниже этой линии, то это будет означать, что портфель показывает очень плохой результат в 5% худших сценариев. Значение используется как оценка VaR (Value at Risk) при 95% уровне доверия.

95-й процентиль:
* Значение может быть использовано для оценки потенциальной прибыли в 5% лучших сценариев. Если симулированное изменение стоимости портфеля находится выше этой линии, это означает, что портфель показывает очень хороший результат в 5% лучших сценариев

Средняя:
* Показывает ожидаемое изменение стоимости портфеля. Это может быть использовано для оценки общей тенденции или тренда изменения стоимости портфеля.

```python
confidence_level = 0.95
# делаем датафрейм
port_pct_change = pd.Series(sim_pct_change[-1,:])

mcVAR = port_pct_change.quantile(1 - confidence_level)
mcCVAR = port_pct_change[port_pct_change <= mcVAR].mean()

# создаем датафрейм заполненный нулями
portfolio_returns = np.full(shape=(T, n_simulation), fill_value=0.0)

# пересчитаем процентные изменения в фактическую стоимость портфеля
for m in range(n_simulation):
    portfolio_returns[:,m] = np.cumprod(sim_pct_change[:,m]+1)*initialPortfolio

# выберем записи последнего смоделированного торгового дня
last_portfolio_returns = portfolio_returns[-1, :]
# VAR и CVAR при 95% доверительном уровне доходности портфеля
mc_var_returns = np.percentile(last_portfolio_returns, 5)
mc_cvar_returns = last_portfolio_returns[last_portfolio_returns <= mc_var_returns].mean()

print(f"Monte Carlo VAR at 95% confidence level: ${mc_var_returns:.2f} ({mcVAR:.2%})")
print(f"Monte Carlo CVAR at 95% confidence level: ${mc_cvar_returns:.2f} ({mcCVAR:.2%})")
```

```python

plt.plot(portfolio_returns)

plt.axhline(mc_var_returns, color='r', linewidth=1, label=f'VAR: ${mc_var_returns:.2f}')
plt.axhline(mc_cvar_returns, color='g', linewidth=1,  label=f'CVAR: ${mc_cvar_returns:.2f}')

plt.legend(loc='upper left')
plt.ylabel('Стоимость портфеля ($)')
plt.xlabel('Дни')
plt.title('Моделирование портфеля методом Монте-Карло')
plt.show()
```
